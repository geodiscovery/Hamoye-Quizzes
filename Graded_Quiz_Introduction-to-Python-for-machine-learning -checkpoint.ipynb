## What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?

```python
# Mean for the year 2015
mean_2015= data['Y2015'].mean()
formatted_mean_2015 = "{:.3f}".format(mean_2015)
print(formatted_mean_2015)
```
